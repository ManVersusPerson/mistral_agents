```python
import ollama
import datasets
import pandas as pd
```

```python
from datasets import load_dataset

dataset = load_dataset("gsm8k", 'main')
```

```python
true_answers = (dataset['train'][:]['answer'][:100])
questions = dataset['train'][:]['question'][:100]
```

```python
answers_1 =[]
answers_2 =[]
answers_3 =[]
```

```python
def ask_question(question):
    x = ollama.chat(model='mistral', messages=[
    {
    'role': 'user',
    'content': question,
    }
    ])
    return x['message']['content']
```

```python
for question in questions:
    
    answer_1 = ask_question(question)
    answers_1.append(answer_1)

    answer_2 = ask_question(question)
    answers_2.append(answer_2)

    answer_3 = ask_question(question)
    answers_3.append(answer_3)
```

```python
answers_df = pd.DataFrame({'answers_1': answers_1, 'answers_2': answers_2, 'answers_3': answers_3, 'true_answers': true_answers})
```
